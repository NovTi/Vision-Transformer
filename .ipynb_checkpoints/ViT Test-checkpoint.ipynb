```python
class PatchEmbedding(nn.Module):
    def __init__(self, in_channel, patch_size, image_size):
        super().__init__()
        self.embed_size = ((image_size // patch_size)**2) * in_channel 
        self.patch_size = patch_size
        self.patch_embed = nn.Sequential(
            nn.Conv2d(in_channel, self.embed_size, kernel_size=self.patch_size, stride=self.patch_size),
            Rearrange('b e (h) (w) -> b (h w) e'),
        )
        self.cls_token = nn.Parameter(torch.rand((1, self.embed_size)))
        self.position_token = nn.Parameter(torch.rand(((image_size//patch_size)**2+1, self.embed_size)))
        
    def forward(self, x):
        b,_,_,_ = x.shape
        out = self.patch_embed(x)
        # concat class token
        cls_tokens = repeat(self.cls_token, 'h e -> b h e', b=b)
        out = torch.cat([cls_tokens, out], dim=1)
        # add position token
        out = out + self.position_token
        
        return out

x = torch.rand((2, 3, 128, 128))
out = PatchEmbedding(3, 32, 128)
out(x).shape
        
```
